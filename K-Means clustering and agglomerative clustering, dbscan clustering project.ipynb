# agglomerative clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram
```

```python
a=AgglomerativeClustering()
p=a.fit_predict(flipkartnumerical[['Rating','Selling Price']])
p
```

```python
u=hierarchy.linkage(a.children_)
u
```

```python
hierarchy.set_link_color_palette(['r','g'])
i=hierarchy.dendrogram(u,truncate_mode='level')
```

# comments:
1. The agglomerative clustering is bottom-up method here in this hierarchy two clustes merged together based on similarities form a new cluster based on euclidean distance.
2. agglomerative clustering is average linkage it calculates average distance of data points between clusters can be calculated.
3. In agglomerative clustering distance is average among one cluster with another cluster in hierarchy.
4. In the above dendrogram 3 main clusters are formed.

```python
a=AgglomerativeClustering(n_clusters=3)
```

```python
s=a.fit_predict(flipkartnumerical[['Rating','Selling Price']])
s
```

```python
flipkartnumerical['Agglomerative Clustering']=s
flipkartnumerical
```

# dbscan 

```python
from sklearn.cluster import DBSCAN
import numpy as np
import seaborn as sns
```

```python
t=[]
for i in np.linspace(0.001,20,50):
    d=DBSCAN(eps=i,min_samples=5)
    d.fit(flipkartnumerical[['Rating','Selling Price']])
    outliers=np.sum(d.labels_==-1)/len(d.labels_)
    t.append(outliers)
```

```python
sns.lineplot(x=np.linspace(0.001,20,50),y=t,color='green',marker='*',markersize=10,markerfacecolor='pink')
```

# so eps 0.2 is good

```python
u=DBSCAN(eps=0.2,min_samples=4)
```

```python
u.fit(flipkartnumerical[['Rating','Selling Price']])
```

```python
b=u.labels_
b
```

```python
flipkartnumerical['DBSCAN']=b
flipkartnumerical
```

# comments:
1. The above dbscan (density-based spatial clustering application with noise) is helps to create cluster for high density data points.
2. dbscan uses three main points these are core points,decision points,eps(epsilon).
3. Core points are created to make a small cluster with decision points and epsilon distance.
4. Decision points are points but not core points but these helps to form a corepoints.
5. epsilon is distance between decisison points and core points will be calculated it should be within epsilon criterion to form small cluster.
6. If decision points and core points are not in epsilon distance again different core points are created with decision points within epsilon still criterion doesnot meet randomly corepoints and decision points are selected to form small clusters.
7. In the above dbscan -1 is outliers and 0 is a decision points where possible to form many core points.
8. Remaining points are clusters these are small clusters formed from density cluster.
